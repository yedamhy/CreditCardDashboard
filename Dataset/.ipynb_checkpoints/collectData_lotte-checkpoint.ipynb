```python
for card_i in range(len(card_num)):
    card_data = {}
    soup = BeautifulSoup(browser.page_source, 'html.parser')

    # 우선 전체 카드 리스트를 확보하고
    for _ in range(4):
        soup = BeautifulSoup(browser.page_source, 'html.parser')
        time.sleep(0.5)
        soup = BeautifulSoup(browser.page_source, 'html.parser')

        body = browser.find_elements('css selector', 'body')[0]
        
        body.send_keys(Keys.PAGE_DOWN)
        body.send_keys(Keys.PAGE_DOWN)
        body.send_keys(Keys.PAGE_DOWN)
        soup = BeautifulSoup(browser.page_source, 'html.parser')

        body.send_keys(Keys.PAGE_DOWN)
        body.send_keys(Keys.PAGE_DOWN)
        body.send_keys(Keys.PAGE_DOWN)

        soup = BeautifulSoup(browser.page_source, 'html.parser')

        try:
            browser.find_element(By.ID, "btnMore").click()
            time.sleep(3)
            soup = BeautifulSoup(browser.page_source, 'html.parser')    

            soup = BeautifulSoup(browser.page_source, 'html.parser')    

            body.send_keys(Keys.PAGE_DOWN)
            body.send_keys(Keys.PAGE_DOWN)
            soup = BeautifulSoup(browser.page_source, 'html.parser')

        except NoSuchElementException:
            soup = BeautifulSoup(browser.page_source, 'html.parser')

            body = browser.find_elements('css selector', 'body')[0]

            body.send_keys(Keys.PAGE_DOWN)
            body.send_keys(Keys.PAGE_DOWN)
            body.send_keys(Keys.PAGE_DOWN)
            soup = BeautifulSoup(browser.page_source, 'html.parser')

            body.send_keys(Keys.PAGE_DOWN)
            body.send_keys(Keys.PAGE_DOWN)
            body.send_keys(Keys.PAGE_DOWN)

            soup = BeautifulSoup(browser.page_source, 'html.parser')

            browser.find_element(By.ID, "btnMore").click()

            soup = BeautifulSoup(browser.page_source, 'html.parser')    

            time.sleep(3)

            soup = BeautifulSoup(browser.page_source, 'html.parser')    

            body.send_keys(Keys.PAGE_DOWN)
            body.send_keys(Keys.PAGE_DOWN)
            soup = BeautifulSoup(browser.page_source, 'html.parser')

            
    # 카드사, 카드 이름
    soup = BeautifulSoup(browser.page_source, 'html.parser')

    card_data['company'] = '롯데카드'
    card_data['title'] = soup.select('#ajaxCardList > li > a > b')[card_i].text



    # 페이지 넘어가서 카드 상세 정보 확인하기
    javascript_command = soup.select('#ajaxCardList > li > a')[card_i]['onclick'].replace('return false;', '')
    browser.execute_script(javascript_command)

    soup = BeautifulSoup(browser.page_source, 'html.parser')
    try:
        # 카드 이미지
        img = soup.select('#container > div.topBox.cardDtlTop > div.inner > div.cardImg > img')[0]['src']
        card_data['img'] = 'https:' + img

        # 주요 혜택 확인하기
        for i in range(5):
            try:
                benfit_select = soup.select('#container > div.inner.cardDtlCont > div.cardBnfTab > div.bnfCont.smry > ul > li')[i]
                card_data[f"benefit_title_{i+1}"] = benfit_select.select('b')[0].text
                card_data[f"benefit_detail_{i+1}_1"] = benfit_select.select('span.svTit')[0].text
                if len( benfit_select.select('span.svTit')) > 1:
                    card_data[f"benefit_detail_{i+1}_2"] = benfit_select.select('span.svTit')[1].text

                if card_data[f"benefit_title_{i+1}"] == card_data[f"benefit_detail_{i+1}_1"]:
                    card_data[f"benefit_detail_{i+1}_1"] = None

            except IndexError:
                break

        # 연회비 확인하기
        soup = BeautifulSoup(browser.page_source, 'html.parser')


        fee_btn = browser.find_elements('css selector', '#annualBtn')
        fee_btn[0].click()
        soup = BeautifulSoup(browser.page_source, 'html.parser')

        for i in range(3):
            try:
                card_data[f"fee_brand_{i+1}"] = soup.select('#layer_annual_fee > div.layCont > div > dl > dt')[i].text.lstrip()

                fee_total_select = soup.select('#layer_annual_fee > div.layCont > div > dl')[i]
                card_data[f"total_fee_{i+1}"] = fee_total_select.select('dd > b')[2].text
            except IndexError:
                card_data[f"fee_brand_{i+1}"] = None
                card_data[f"total_fee_{i+1}"] = None

        time.sleep(0.5)

        browser.find_elements('css selector', '#layer_annual_fee > button')[0].click()
        soup = BeautifulSoup(browser.page_source, 'html.parser')


        # 카드 출시일 확인하기
        try:
            card_data['date'] = soup.select('#form1 > div > div.inner > div.bnfDtList > ul.listDep1.marT30 > li')[-1].text.replace('카드출시일:','')

        except IndexError:
            try:
                card_data['date'] = soup.select('#form1 > div > div.inner > ul.listDep1.mt30 > li')[-1].text.replace('카드출시일:','')

            except:
                card_data['date'] = soup.select('#container > div.inner.cardDtlCont > ul.listDep1.marT30 > li')[-1].text.replace('카드출시일:','')
        # 카드 상세페이지 url
        card_data['url'] = browser.current_url

        data_list.append(card_data)

        print(card_i+1, card_data['title'])

    except IndexError:
        card_data['index'] = card_i
        except_card.append(card_data)
        
    browser.back()
    soup = BeautifulSoup(browser.page_source, 'html.parser')
```

```python
import re

# 문자열에서 날짜를 추출하는 정규식 패턴 (한글 월과 일 포함)
date_pattern = r'(\d{4})년 (\d{2})월 (\d{2})일'  # 예: "2021년 05월 11일" 형식의 날짜

# 'date' 열에서 모든 날짜를 추출하여 수정합니다.
filtered_df['date'] = filtered_df['date'].str.extract(date_pattern).apply(lambda x: f"{x[0]}년 {x[1]}월 {x[2]}일", axis=1)
filtered_df
```
